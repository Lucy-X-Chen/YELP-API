```python
#this is the link to request search based on multiple parameters
url='https://api.yelp.com/v3/businesses/search'

params = {'term':'french','location':'New York City','price':2, 'limit':5} 
#We can set parameters based on yelp criterias, and then pass the parameters in a dictionary

#parameters includes:
#term:Optional. Search term, for example "food" or "restaurants". The term may also be business names, such as "Starbucks".

#location: Required if either latitude or longitude is not provided. 
#This string indicates the geographic area to be used when searching for businesses. 
#Examples: "New York City", "NYC", "350 5th Ave, New York, NY 10118". 

#Categories:  For example, "bars,french" will filter by Bars OR French.

#limit:Number of business results to return. By default, it will return 20. Maximum is 50.

#price:Pricing levels to filter the search result with: 1 = $, 2 = $$, 3 = $$$, 4 = $$$$. 

#sort_by: Suggestion to the search algorithm that the results be sorted by one of the these modes: 
#best_match, rating, review_count or distance. The default is best_match. 

#In total there are 14 parameters can be sets, more terms to be find https://www.yelp.com/developers/graphql/query/search

# Making a get request to the API
req1 = requests.get(url, params=params, headers=headers)
 
# proceed only if the status code is 200
print('The status code is {}'.format(req1.status_code))

#200 OK status means that your request was successful,
#whereas a 404 NOT FOUND status means that the resource you were looking for was not found

#we get the request text by using json.loads
json.loads(req1.text)


```

```python
#Parse json data and print in nice readable format

url = 'https://api.yelp.com/v3/businesses/search'
params = {'term':'french','location':'New York City','price':2, 'limit':5} 
req3 = requests.get(url, params=params, headers=headers)
parsed = json.loads(req3.text)

#all the objects stored and represent yelp business: https://www.yelp.com/developers/graphql/objects/business

businesses = parsed["businesses"]
 
for business in businesses:
    print("Name:", business["name"])
    print("Rating:", business["rating"])
    print("Address:", " ".join(business["location"]["display_address"]))
    print("Phone:", business["phone"])
    print("\n")
 
    id = business["id"]
 
    url="https://api.yelp.com/v3/businesses/" + id + "/reviews"
    
    req3 = requests.get(url, headers=headers)
 
    parsed = json.loads(req3.text)
 
    reviews = parsed["reviews"]
 
    print("--- Reviews ---")
 
    for review in reviews:
        print("User:", review["user"]["name"], "Rating:", review["rating"], "Review:", review["text"], "\n")
 
```
